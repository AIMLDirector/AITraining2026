```python
import matplotlib.pyplot as plt
df['Salary'].plot(kind='line')
plt.title("Line Plot of Column Salary")
plt.xlabel("Salary")
plt.ylabel("Value")
plt.show()
```

```python
min_salary = df['Salary'].min()
max_salary = df['Salary'].max()
salary_range = max_salary - min_salary
mean_salary = df['Salary'].mean()

# Format the text string
summary_text = (f"Summary Statistics:\n"
                f"Mean: ${mean_salary:,.2f}\n"
                f"Min: ${min_salary:,.2f}\n"
                f"Max: ${max_salary:,.2f}\n"
                f"Range: ${salary_range:,.2f}")

# Plot the data (e.g., a histogram for distribution)
plt.hist(df['Salary'], bins=10, color='skyblue', edgecolor='black')
plt.title("Salary Distribution with Summary")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")

# Add the summary text to the plot
# Using transform=plt.gca().transAxes places the text relative to the axes (0,0 is bottom-left, 1,1 is top-right)
plt.text(0.95, 0.95, summary_text, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='right',
         bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.7)) # Add a text box for clarity

plt.show()
```

```python
import pandas as pd
nba = pd.read_csv("nba.csv").dropna(how = "all")

nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
```

```python
nba.tail()
```

```python
nba.dtypes

nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")

```

```python
nba.sort_values("Name", ascending= False)
nba.sort_values("Salary", ascending= False)

nba.sort_index(ascending= False)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
```
